**1. Import Libraries**

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

**Step 2 — Load Dataset**





```python
data = pd.read_csv("household_power_consumption.txt", sep=';', low_memory=False)
```

**Step 3 — Data Cleaning**

```python
data = data.replace('?', np.nan)
data = data.dropna()

# Convert to float
for col in data.columns[2:]:
    data[col] = data[col].astype(float)
```

**Step 4 — Select Features**

```python
X = data[['Voltage', 'Global_intensity']]
y = data['Global_active_power']
```

**Step 5 — Without Scaling (Baseline Model)**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse_without = mean_squared_error(y_test, y_pred)
print("MSE without scaling:", mse_without)
```

**Step 6 — Apply StandardScaler**

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

**Step 7 — Train Model After Scaling**

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse_with = mean_squared_error(y_test, y_pred)
print("MSE with scaling:", mse_with)
```
